# Year

```python
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
sns.distplot(df['Year'])
plt.show()
```

## Observation

- Year is Left skewed.
- It means there are presence some outliers.

```python
df['Year'].describe()
```

# Finding the IQR

```python
percentile25 = df['Year'].quantile(0.25)
percentile75 = df['Year'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit", upper_limit)
print("Lower limit", lower_limit)
```

## Observation

- The outliers only at upper limit because the values are far from 0.

# Finding outliers

```python
df[df['Year'] > upper_limit]
```

```python
df[df['Year'] < lower_limit]
```

# Capping

```python
new_df_cap = df.copy()

new_df_cap['Year'] = np.where(
    new_df_cap['Year'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Year'] < lower_limit,
        lower_limit,
        new_df_cap['Year']
    )
)
```

```python
new_df_cap.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Year'])

plt.subplot(2,2,2)
sns.boxplot(df['Year'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['Year'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Year'])

plt.show()
```
