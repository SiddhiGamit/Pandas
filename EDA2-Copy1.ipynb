Task:
date of cell, price, cell count, product name, (10 data)
1.price = $ 5-10 values, not null
2.product count = 5, 000    10,000 not missing
3.product name = 7/10 missing/name p1 p2 p3
4.date of cell = 10  not missing
5.sale 3 missing (int)

```python
data = {
    'price' : ['$1000', '$1220', '$1300', '$1456', '$1674', '$352', '$2454', '$5232', '$1289', '$7654'],
    'product count' : ['5,000', '7,100', '5,874', '85,452', '14,725', '5,478', '89,741', '10,000', '77,000', '1,000'],
    'product name' : ['p1', 'p3', np.nan, 'p2', np.nan, 'p3', 'p3', 'p2', np.nan, 'p1'],
    'date of sale' : ['04/01/2001', '25/02/2002', '09/03/2003', '30/04/2004', '14/05/2005', '28/06/2006', '25/07/2007', '17/08/2008', '09/09/2009', '10/10/2010'],
    'sale' : [77, 55, 71, np.nan, 23, np.nan, 33, 58, 66, np.nan]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
def filter_product_count(x):
    x = float(x.replace(',',''));
    return x

df['product count'] = df['product count'].apply(filter_product_count)
```

```python
df
```

```python
def filter_price(x):
    x = float(x.replace('$',''));
    return x

df["price"] = df["price"].apply(filter_price)
```

```python
df
```

```python
df['date of sale'] = pd.to_datetime(df['date of sale'])
df
```

```python
df['sale'] = df['sale'].fillna(df['sale'].mean())
df
```

```python
df["product name"].fillna(method="ffill" , inplace = True)
df
```

```python
df['sale'].fillna(int(df['sale'].median()), inplace = True)
df
```
