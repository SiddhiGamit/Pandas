```python
df['Rating'] = df['Rating'].astype(float)
df.dtypes
```

```python
df['Size'] = df['Size'].astype(float)
df.dtypes
```

```python
df['Installs'] = df['Installs'].astype(float)
df.dtypes
```

```python
def filter_price(x):
    x = float(x.replace("$",""))
    return x

df["Price"] = df["Price"].apply(filter_price)
```
